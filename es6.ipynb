```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs

#utiliziamo la funzione make_blobs di scikit-learn per creare un dataset di 500 esempi
X, y_true = make_blobs(n_samples = 1200, centers = 7, cluster_std= 1.8)

plt.scatter(X[:, 0], X[:,1])
```

```python
from sklearn.cluster import KMeans

#
kmeans = KMeans(n_clusters= 7)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
#X[:,0] accede alla prima colonna(colona 0) dei dati all'interno di df_cluster
plt.scatter(X[:, 0], X[:, 1], c = y_kmeans, s = 50, cmap = 'viridis')

centers = kmeans.cluster_centers_

plt.scatter(centers[:,0], centers[:,1], c = 'red', s = 200, alpha=0.5)
```

re il numero di cluster nell esempio abbiamo il numero di clusters a 4

```python
sse ={}

for k in range(1, 10):
  kmeans = KMeans(init= 'k-means++', n_clusters= k).fit(X)
  sse[k] = kmeans.inertia_ #Inertia: Sum of distances of samples to theircloses

plt.figure()
plt.plot(list(sse.keys()), list(sse.values()), marker = 'o')
plt.xlabel('Numero di clusters', fontsize = 16)
plt.ylabel('Somma della distanza al quadrato', fontsize = 16)
plt.savefig('number_of_k.png') #salva una foto del grafico con il nome
plt.show()
```
